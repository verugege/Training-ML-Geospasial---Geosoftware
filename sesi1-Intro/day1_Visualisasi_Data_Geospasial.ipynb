**Import Library**

```python
import geopandas as gpd #untuk mengelola data geospasial
import matplotlib.pyplot as plt #visualisasi data geospasial
```

**Input lokasi data**

```python
shp_path = '/content/drive/MyDrive/training_ml_geospasial_geosoftware/sesi1/ADMINISTRASIKECAMATAN_AR_25K/ADMINISTRASIKECAMATAN_AR_25K.shp'
```

```python
## Load Data
gdf = gpd.read_file(shp_path)
```

```python
## Check Data
gdf
```

```python
gdf.info()
```

```python
## Visualisasi Data Geospasial

## membuat plot peta
gdf.plot(
    figsize = (10,10),
    edgecolor = "black",
    cmap = 'Set3'
)

## membuat judul peta
plt.title('Peta Administrasi Jakarta Pusat')

## Menambahkan Label Peta
for index, row in gdf.iterrows():
  x=row['geometry'].centroid.x
  y=row['geometry'].centroid.y

  label = f"{row['NAMOBJ']}\nArea: {row['SHAPE_Area']} ha"
  plt.text(x,y, label, fontsize=8, ha='center')
  #plt.text(x,y, row['NAMOBJ'], fontsize=8, ha='center')

# menampilkan posisi peta
plt.axis('on')

# tampilkan gambar
plt.show
```
